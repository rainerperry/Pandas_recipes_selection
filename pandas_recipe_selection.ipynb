```python
#The path to the CSV files

file1 = "Resources/recipes.csv"
file2 = "Resources/reviews.csv"
```

```python
#read CSV files

recipes_df = pd.read_csv(file1)
reviews_df = pd.read_csv(file2)
```

```python
#get unique values of "RecipeCategory" column to be able to select desired categories

#print(recipes_df.RecipeCategory.unique())

recipes_df['RecipeCategory'].value_counts()
```

```python
#extract only those with category "dessert"

dessert_df = recipes_df.loc[(recipes_df["RecipeCategory"] == 'Dessert')]
dessert_df.head()
dessert_df.shape[0]
```

```python
#left join with reviews_df

merged_dessert = pd.merge(dessert_df, reviews_df, how='left',on=['RecipeId', 'AuthorId', 'AuthorName'])

merged_dessert.head()
dessert_df.shape[0]
```

```python
#get only those from last 2 years in 2 steps: 1) determine range of dates, 2) select top 24 months

print(min(merged_dessert['DatePublished']))
print(max(merged_dessert['DatePublished']))
```

```python
# convert datetime to date

merged_dessert['DatePublished'] = pd.to_datetime(merged_dessert['DatePublished'])

merged_dessert.head(5)
```

```python
merged_dessert.count()
```

```python
#searchign for 2020 dates only

start_date = '2020-01-01'
end_date= '2020-12-22'

date_range = (merged_dessert['DatePublished'] >= start_date) & (merged_dessert['DatePublished'] <= end_date)


```

```python
twenty_only = merged_dessert.loc[date_range]
```

```python
twenty_only.head()
```

```python
twenty_only.count()
```

```python
twenty_only['Keywords'].value_counts(ascending=False)
```

```python
thirty_min = twenty_only[twenty_only['Keywords'].str.contains('< 30 Mins', na=False)]

#note: "na=False" is necessary to tansform NA values and avoid error message

thirty_min.head()
```

```python
thirty_min.info()
```

```python
clean_thirty = thirty_min.drop(columns=['AggregatedRating', 'ReviewCount', 
                                        'ReviewId', 'Rating', 'Review', 'DateSubmitted', 'DateModified'])
                                        
                               
```

```python
clean_thirty.info()
```

```python
clean_thirty = clean_thirty.dropna(how='any')
```
